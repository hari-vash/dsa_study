# Arrays

```python
arr = [1,2,3]
print(arr)
```

```python
# append: on average: O(1)
arr.append(4)
print(arr)
```

```python
# Pop - O(1)
arr.pop()
print(arr)
```

```python
# insert(not at the end) - O(n)
arr.insert(2,4)
print(arr)
```

```python
# modify - O(1)
arr[0] = 7
print(arr)
```

```python
# accessing an element given index i - O(1)
print(arr[3])
```

```python
# checking if array has an element - O(n)
if 7 in arr:
    print(True)
```

```python
# checking length - O(1)
print(len(arr))
```

# Strings

```python
# append (end of string) - O(n)
s = "hello"
b = s + "z"
print(b)
```

```python
# check if something in string - O(n)
if "f" in s:
    print(True)
```

```python
# access positions - O(1)
print(s[2])
```

```python
# check length of string - O(1)
len(s)
```
